```python
# Importing all required libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Reading the data
df = pd.read_csv("C:/Users/mahat/Portfolio Projects/Youtube Stats Analysis/GlobalYouTubeStatistics.csv")
df.columns
```

# Top 10 Subscribed Channels on Youtube

```python
from matplotlib.ticker import FuncFormatter

#Formatting subscribers for millions
def millions(x, pos):
    return f'{x* 1e-6: .0f}M'

# Calculating top channels with highest subscribers
top_channels = df.sort_values(by='subscribers', ascending=False).head(10)

# Visualization
plt.figure(figsize = (12,6))
plt.bar(top_channels['Youtuber'],top_channels['subscribers'], color = 'green')
plt.ylabel('Subscribers Count')
plt.title('Top 10 Subscribed Channels on Youtube')
plt.xticks(rotation=45)
plt.gca().yaxis.set_major_formatter(FuncFormatter(millions))
plt.show()
```

# Top 10 Viewed Channels on Youtube

```python
# Formatting views for billions
def billions(x,pos):
    return f'{x * 1e-9:.2f}B'

# Calculating top channels with highest views
top_channels = df.sort_values(by='video views', ascending=False).head(10)

# Visualization
plt.figure(figsize = (12,6))
plt.bar(top_channels['Youtuber'],top_channels['video views'], color = 'blue')
plt.ylabel('Video Views')
plt.title('Top 10 Viewed Channels on Youtube')
plt.xticks(rotation=45)
plt.gca().yaxis.set_major_formatter(FuncFormatter(billions))
plt.show()
```

# Uploads vs Subscribers

```python
# Data
UploadCount = df['uploads']
SubscribersCount = df['subscribers']

plt.figure(figsize=(12, 6))

# Visualizing the relationship with Scatter plot
plt.scatter(
    UploadCount, SubscribersCount,
    c='mediumseagreen',      
    s=40,                     
    alpha=0.6,                 
    edgecolors='black'   
)
# Formatting upload counts for thousands
def thousands(x, pos):
    return f'{x * 1e-3:.0f}K'
    
#Formatting the subscriber count to millions and upload counts to thousands
plt.gca().xaxis.set_major_formatter(FuncFormatter(thousands))
plt.gca().yaxis.set_major_formatter(FuncFormatter(millions))

# Adding a grid
plt.grid(True, linestyle='--', alpha=0.3)

# Titles and labels with formatting
plt.title("Uploads vs Subscribers", fontsize=16, fontweight='bold')
plt.xlabel("Number of Uploads", fontsize=12)
plt.ylabel("Number of Subscribers", fontsize=12)

# Tight layout for spacing
plt.tight_layout()

# Show plot
plt.show()
```

# Subscribers and Highest earnings (Yearly)

```python
# Data
SubscriberCount = df['subscribers']
Earnings = df['highest_yearly_earnings']

plt.figure(figsize=(12, 6))

# Visualizing the relationship with Scatter plot
plt.scatter(
    SubscriberCount, Earnings,
    c='red',      
    s=40,                     
    alpha=0.6,                 
    edgecolors='black'   
)
    
#Formatting the subscriber count to millions and earnings to billions
plt.gca().xaxis.set_major_formatter(FuncFormatter(millions))
plt.gca().yaxis.set_major_formatter(FuncFormatter(billions))

# Adding a grid
plt.grid(True, linestyle='--', alpha=0.3)

# Titles and labels with formatting
plt.title("Subscribers and Earnings", fontsize=16, fontweight='bold')
plt.xlabel("Subscribers Count", fontsize=12)
plt.ylabel("Earnings", fontsize=12)

# Tight layout for spacing
plt.tight_layout()

# Show plot
plt.show()
```
